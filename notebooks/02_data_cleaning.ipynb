# Retail Customer Analytics & Prediction

## Data Cleaning

Objective:
Clean and transform raw retail transaction data by removing duplicates, handling missing values, correcting data types, and creating new analytical features.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/raw/OnlineRetail.csv",
    encoding="unicode_escape"
)

df.head()
```

```python
df.shape
```

## Removing Duplicate Transactions

```python
df = df.drop_duplicates()

df.shape
```

## Convert Date Column

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

df.info()
```

## Revenue Calculation

Revenue is calculated as:

Revenue = Quantity × UnitPrice

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]

df.head()
```

```python
df.tail()
```

```python
df.to_csv(
    "../data/cleaned/retail_cleaned.csv",
    index=False
)
```
